```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
db_norm=norm_func(db.iloc[:,1:])
db_norm
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram=sch.dendrogram(sch.linkage(db_norm,method='average'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(db_norm)
y_hc
```

```python
db['h_clusterid']=hc.labels_
db
```

```python
#by K_means
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
db1=pd.read_csv('crime_data.csv')
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
db_norm=norm_func(db.iloc[:,1:])
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(db_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model=KMeans(n_clusters=4)
model.fit(db_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
db1['Clust']=x
```

```python
db1.iloc[:,1:5].groupby(db1.Clust).mean()
```
